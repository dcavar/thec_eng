```python
import spacy
from spacy import displacy
import svgling
```

```python
spacy.prefer_gpu()
nlp = spacy.load("en_core_web_trf")
```

```python
doc = nlp("Paul and John were watching the news, and Mary a movie.")
```

```python
displacy.render(doc, style="dep", options = {'distance':90} )
```

```python
doc = nlp("Paul and John were watching the news, and Mary was watching a movie.")
```

```python
displacy.render(doc, style="dep", options = {'distance':90} )
```
